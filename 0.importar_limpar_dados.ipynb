# Leitura dos dados

```python
import numpy as np
import pandas as pd
import os
from datetime import date, timedelta, datetime

pd.set_option('display.max_columns', 45)
```

```python
wpath = os.getcwd()
path = f'{wpath}/dados/linha_150'
pastas = os.listdir(path)
print(pastas)
```

```python
dadosXls = []

for p in pastas:
    arquivos = os.listdir(f'{path}/{p}')
    for x in arquivos:
        if '.xlsx' in x:
            data = pd.read_excel(f'{path}/{p}/{x}')
            dadosXls.append(data)

dadosXls = pd.concat(dadosXls)
dadosXls.head()
```

```python
dadosXls['Unnamed: 3'] = dadosXls['Unnamed: 3'].fillna(method="ffill")

dadosXls['Unnamed: 1'].where(
    dadosXls['Unnamed: 1'].str.contains("Data: "), inplace=True)

dadosXls['Unnamed: 1'] = dadosXls['Unnamed: 1'].fillna(method="ffill")
```

```python
dados = dadosXls.loc[(dadosXls['Unnamed: 5'].isin(['IDA', 'VOLTA'])) & (
    dadosXls['Unnamed: 7'] != '-') & (dadosXls['Unnamed: 11'] != '-')].reset_index()
```

```python
dados.head(3)
```

```python
dados.loc[:,'data'] = pd.to_datetime(
    dados['Unnamed: 1'].str.slice(6, 16), format='%d/%m/%Y')

dados.loc[:,'linha'] = dados['Unnamed: 1'].str.slice(26, 29)
```

```python
# horário inicio
dados.loc[:, 'inicior'] = dados['Unnamed: 1'].str[12: 16]+'-' \
    + dados['Unnamed: 1'].str[9: 11]+'-' + \
    dados['Unnamed: 1'].str[6: 8]+' ' + \
    dados['Unnamed: 7'].str[6: 8] + dados['Unnamed: 7'].str[9: 11]

dados.loc[:, 'inicior'] = pd.to_datetime(dados.loc[:, 'inicior'])
```

```python
# horário término
dados.loc[:, 'terminor'] = dados['Unnamed: 1'].str[12: 16]+'-' \
    + dados['Unnamed: 1'].str[9: 11]+'-' + \
    dados['Unnamed: 1'].str[6: 8]+' ' + \
    dados['Unnamed: 11'].str[6: 8] + dados['Unnamed: 11'].str[9: 11]

dados.loc[:, 'terminor'] = pd.to_datetime(dados.loc[:, 'terminor'])
```

```python
dados.loc[:, 'servico'] = dados['Unnamed: 3'].str[8:]
dados.loc[:, 'paradas_cumpridas'] = dados['Unnamed: 20'].str.split(',').str[0]
```

```python
dados_filtered = dados[['Unnamed: 14', 'Unnamed: 15',
                        'Unnamed: 16', 'Unnamed: 17', 'data',	'linha',	'inicior',	'terminor']]
```

```python
dados_filtered.head()
```

```python
dados_filtered = dados_filtered.rename(columns={
    'Unnamed: 14': 'ponto_inicial',
    'Unnamed: 15': 'ponto_final',
    'Unnamed: 16': 'servico',
    'Unnamed: 17': 'trajeto'
})
```

```python
dados_filtered.columns
```

## exportar CSV

```python
dados_filtered[['linha', 'data', 'trajeto', 'ponto_inicial', 'ponto_final',
                'servico', 'inicior', 'terminor']].to_csv('linha150.csv', index=False)
```
